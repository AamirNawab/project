```python
# function to read CSV and excel files
def read_data(file_url):
    try:
        data = pd.read_csv(file_url)
    except:
        data = pd.read_excel(file_url)
        return(data)
```

```python
import wget
import zipfile
import os.path
```

```python
#Function to Unzip Files
path = "C:\\Users\\aamir\\OneDrive\\Desktop\\HACE_Project\\zip_files"
import os

def unzip(url):
    zip_file = wget.download(url)
    with zipfile.ZipFile(zip_file, 'r') as zip_obj:
        zip_obj.extractall(path)
    logdir= path # path to directory
    logfiles = sorted([ f for f in os.listdir(logdir) if f.endswith('.csv')])
    data = read_data(path + "\\"+logfiles[-1],)
    return(data)
```
